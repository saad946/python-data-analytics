```python
import matplotlib.pyplot as plt
import numpy as np

# List of data points
l = [25, 46, 57, 25, 45, 13, 29, 39, 46, 99, 22, 12, 14, 16, 55, 6, 46, 52, 55]

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = np.percentile(l, 25)
Q3 = np.percentile(l, 75)

# Calculate IQR
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
l_no_outliers = [x for x in l if lower_bound <= x <= upper_bound]

# Create boxplot without outliers
plt.boxplot(l_no_outliers)

# Display the plot
plt.title("Boxplot Without Outliers")
plt.show()

# Print original and filtered lists
print("Original data:", l)
print("Data without outliers:", l_no_outliers)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the expense data
expense_data = pd.read_excel("C:/Users/Dell/Desktop/Python/Python Data Analytics/datasets/expense3.xlsx")
df = pd.DataFrame(expense_data)

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = expense_data["Amount"].quantile(0.25)
Q3 = expense_data["Amount"].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data to remove outliers
filtered_expense_data = expense_data[(expense_data["Amount"] >= lower_bound) & (expense_data["Amount"] <= upper_bound)]

# Display boxplot before and after removing outliers
plt.figure(figsize=(12, 6))

# Original data boxplot
plt.subplot(1, 2, 1)
plt.boxplot(expense_data["Amount"])
plt.title("Boxplot with Outliers")

# Filtered data boxplot
plt.subplot(1, 2, 2)
plt.boxplot(filtered_expense_data["Amount"])
plt.title("Boxplot without Outliers")

plt.show()

# Display the filtered data
print("Filtered Expense Data (Without Outliers):")
print(filtered_expense_data)
```
